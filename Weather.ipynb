```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
sns.set(color_codes=True)
```

```python
weather=pd.read_csv('data/Test.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
# Create a rainbow color palette
num_bars = len(weather['humidity'].unique())
colors = plt.cm.rainbow(np.linspace(0, 1, num_bars))

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='humidity', y='temperature', data=weather, palette=colors)

# Show the plot
plt.title('Temperature by Humidity')
plt.xlabel('Humidity')
plt.ylabel('Temperature')
plt.show()
```

```python
sns.displot(weather['humidity'])
plt.title('Frequency by Humidity')
```
